```python
# Compound Interest
P = float(input("Enter Principal: "))
R = float(input("Enter Rate (decimal e.g. 0.05 for 5%): "))
n = int(input("Enter number of times compounded per year: "))
t = float(input("Enter Time (years): "))

A = P * (1 + R/n) ** (n * t)

print("Amount (Compound Interest) =", A)
```
