```python
import numpy as np
import math

print("Equation Solver")
print("1. Quadratic Equation")
print("2. Cubic Equation")
print("3. Quartic Equation")

choice = int(input("Select option (1-3): "))

# Quadratic
if choice == 1:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))

    d = b**2 - 4*a*c

    if d >= 0:
        x1 = (-b + math.sqrt(d)) / (2*a)
        x2 = (-b - math.sqrt(d)) / (2*a)
    else:
        x1 = (-b + complex(0, math.sqrt(-d))) / (2*a)
        x2 = (-b - complex(0, math.sqrt(-d))) / (2*a)

    print("Roots:", x1, x2)

# Cubic
elif choice == 2:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))

    roots = np.roots([a, b, c, d])
    print("Roots:", roots)

# Quartic
elif choice == 3:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    c = float(input("Enter c: "))
    d = float(input("Enter d: "))
    e = float(input("Enter e: "))

    roots = np.roots([a, b, c, d, e])
    print("Roots:", roots)

else:
    print("Invalid choice")
```
