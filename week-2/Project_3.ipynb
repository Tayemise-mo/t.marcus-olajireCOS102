```python
from tabulate import tabulate

data = [["Eve", "Class A", "Section 1", "85%"],
        ["Jane Smith", "Class B", "Section 2", "92%"],
        ["Peter Jones", "Class A", "Section 3", "78%"]]

headers = ["Student Name", "Class", "Section", "Percentage"]

# Print the table with a grid format
print(tabulate(data, headers=headers, tablefmt="grid"))
```
