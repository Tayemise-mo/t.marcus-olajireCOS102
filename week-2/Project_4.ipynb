```python
years_of_experience = int(input("Your Years of Experience please: "))
age = int(input("Enter your age please: "))
          
if years_of_experience > 25 and age >= 55:
          print("Your Annual Tax Revenue is N5_600_000")
elif years_of_experience > 20 and age >= 45:
          print("Your Annual Tax Revenue is N4_480_000")
elif years_of_experience > 10 and age >= 35:
          print("Your Annual Tax Revenue is N1_500_000")
elif years_of_experience < 10 and age < 35:
          print("Your Annual Tax Revenue is N550_000")
else:
    print("UNAVAILABLE")
```
